Este notebook analiza el dataset sintético generado para el clasificador de documentos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')
from pathlib import Path
from collections import Counter

# Cargar ground truth
gt = pd.read_csv('../data/ground_truth.csv')
print('Total documentos:', len(gt))
print(gt.head())
```

```python
# Distribución de clases
dist = gt['categoria'].value_counts()
print('Distribución por categoría:')
print(dist)

fig, ax = plt.subplots(figsize=(10, 5))
dist.plot(kind='bar', ax=ax, color='steelblue', edgecolor='black')
ax.set_title('Distribución de Documentos por Categoría')
ax.set_xlabel('Categoría')
ax.set_ylabel('Cantidad de Documentos')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.savefig('eda_distribucion_clases.png', dpi=100, bbox_inches='tight')
print('Gráfica guardada: eda_distribucion_clases.png')
```

```python
# Análisis de longitud de documentos
print('Estadísticas de longitud (palabras):')
print(gt.groupby('categoria')['num_palabras'].describe())

fig, ax = plt.subplots(figsize=(10, 5))
gt.boxplot(column='num_palabras', by='categoria', ax=ax)
ax.set_title('Distribución de Longitud por Categoría')
ax.set_xlabel('Categoría')
ax.set_ylabel('Número de Palabras')
plt.suptitle('')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('eda_longitud_documentos.png', dpi=100, bbox_inches='tight')
print('Gráfica guardada.')
```

```python
# Vocabulario clave por categoría (top 10 palabras)
import re

stopwords_es = {'de', 'la', 'el', 'en', 'y', 'a', 'que', 'se', 'los',
                'las', 'un', 'una', 'por', 'con', 'es', 'del', 'no',
                'al', 'su', 'lo', 'para', 'como', 'más', 'sus'}

print('Top 10 palabras por categoría:')
for cat in gt['categoria'].unique():
    archivos = gt[gt['categoria'] == cat]['archivo'].tolist()
    textos = []
    for f in archivos[:10]:  # Primeros 10 archivos
        try:
            texto = Path(f).read_text(encoding='utf-8').lower()
            palabras = re.findall(r'[a-záéíóúñ]{4,}', texto)
            textos.extend(palabras)
        except:
            pass
    top = Counter(w for w in textos if w not in stopwords_es).most_common(10)
    print(f'  {cat}: {[w for w, c in top]}')
```
